```python
import matplotlib.pyplot as plt
```

```python
survival = data['Survived'].value_counts()
plt.figure(figsize=(6,5))
plt.bar(['Not Survived', 'Survived'], survival)
plt.title("Passenger Survival Count")
plt.xlabel("Survival Status")
plt.ylabel("Number of Passengers")
plt.show()
```

### Insights

- The dataset contains both survivors and non-survivors.
- The number of passengers who did not survive is higher than the number who survived.
- The survival rate indicates that the disaster had a significant impact on passengers.

```python
plt.figure(figsize=(8,5))
plt.hist(data['Age'], bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Insights

- Most passengers were between 20 and 40 years old.
- The number of elderly passengers was relatively low.
- The age distribution suggests that the majority of travelers were young and middle-aged adults.

```python
plt.figure(figsize=(8,5))
plt.hist(data['Fare'], bins=20, edgecolor='black')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

### Insights

- Most passengers paid lower fares.
- Only a small number of passengers paid very high fares.
- The fare distribution is right-skewed, indicating a few high-value ticket purchases.

```python
survival_gender = data.groupby(['Sex', 'Survived']).size().unstack()
survival_gender.plot(kind='bar', figsize=(7,5))
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

### Insights

- Female passengers had a higher survival rate than male passengers.
- Male passengers experienced a greater number of fatalities.
- Gender appears to have had a strong influence on survival.

```python
survival_class = data.groupby(['Pclass', 'Survived']).size().unstack()
survival_class.plot(kind='bar', figsize=(7,5))
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

### Insights

- First Class passengers had the highest survival rate.
- Third Class passengers had the lowest survival rate.
- Passenger class appears to have significantly influenced survival chances.

```python
plt.figure(figsize=(8,6))
plt.scatter(data['Age'], data['Fare'])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

### Insights

- There is no strong linear relationship between age and fare.
- Higher fares were paid by passengers of different age groups.
- Fare appears to be influenced more by passenger class than by age.

```python
import numpy as np

corr = data.corr(numeric_only=True)

plt.figure(figsize=(8,6))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()

plt.xticks(np.arange(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(np.arange(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.show()
```

### Insights

- Fare shows a moderate relationship with passenger class.
- Age has a weak correlation with survival.
- Most numerical features have weak to moderate correlations, indicating that survival was influenced by multiple factors rather than a single variable.

# Conclusion

The Exploratory Data Analysis (EDA) of the Titanic dataset provided valuable insights into passenger demographics and survival patterns. Statistical summaries and visualizations revealed that gender and passenger class had a strong influence on survival, while age had a weaker relationship. Most passengers traveled in Third Class, were between 20 and 40 years old, and embarked from Southampton. Overall, this analysis demonstrates how EDA can uncover meaningful trends and relationships, forming a strong foundation for further predictive modeling and data-driven decision-making.
